# **Clean text data**

```python
# Import libraries
import pandas as pd
import re
import string
from IPython.display import display

# Load datasets
df = pd.read_csv('/content/drive/MyDrive/วิจัย/insult.csv')
df = pd.read_csv('/content/drive/MyDrive/วิจัย/sexual.csv')
df = pd.read_csv('/content/drive/MyDrive/วิจัย/Social_Differentiation.csv')

# Text cleaning function
def clean_msg(msg):
    # Check if input is string
    if not isinstance(msg, str):
        return ""

    # Remove URLs
    msg = re.sub(r'https?://\S+|www\.\S+', '', msg)

    # Remove mentions (@username)
    msg = re.sub(r'@\w+', '', msg)

    # Remove hashtag symbol but keep the word
    # Example: #hate -> hate
    msg = re.sub(r'#([^\s#]+)', r'\1', msg)

    # Remove Thai repetition mark (ๆ)
    msg = msg.replace("ๆ", "")

    # Remove numbers
    msg = re.sub(r'\d+', '', msg)

    # Remove emojis and special symbols
    msg = re.sub(r'[^\x00-\x7F\u0E00-\u0E7F0-9A-Za-z ]+', '', msg)

    # Remove English punctuation
    msg = msg.translate(str.maketrans('', '', string.punctuation))

    # Remove extra spaces
    msg = ' '.join(msg.split())
    return msg


# Apply cleaning function to insult dataset
df_insult['Cleaned_Text'] = df_insult['Text'].apply(clean_msg)
# Apply cleaning function to sexual dataset
df_sexual['Cleaned_Text'] = df_sexual['Text'].apply(clean_msg)
# Apply cleaning function to social differentiation dataset
df_Social_Differentiation['Cleaned_Text'] = df_Social_Differentiation['Text'].apply(clean_msg)

# Display cleaned results
print("===== INSULT =====")
display(df_insult[['Text', 'Cleaned_Text']].head(1000))

print("===== SEXUAL =====")
display(df_sexual[['Text', 'Cleaned_Text']].head(1000))

print("===== SOCIAL DIFFERENCE =====")
display(df_Social_Differentiation[['Text', 'Cleaned_Text']].head(1000))
```

```python
# Import libraries for Thai NLP
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus.common import thai_stopwords
from pythainlp.corpus.common import thai_words
from pythainlp.util.trie import Trie

# Thai stopwords
th_stop = set(thai_stopwords())

# English stopwords
en_stop = set()

# Custom words that should stay as one token
custom_words = [
    "แม่ง", "ชิบหาย", "อมควย", "ส้นตีน",
    "ขอดมหี", "เลียหี", "แตกใน", "ช่วยตัวเอง", "จับกด","เหมน",
    "แว้น", "สถุน","อุบาด","พรรคส้ม","ฝรั่งเศษ","ไร้สมอง","ธรรมนัส","เย้สเขร้",
    "ตลาดล่าง","แอคหลุม","ต่ำตม","เสล่อ","ไร้การศึกษา","คอนเท้นท์","กักขระ","ประเทศไทย","ชาติชั่ว",
    "ภูมิใจไทย","อิดอกทอง","เเป้ง","ชิปฝาย","เมกา","เป้น","แปป",
    "ฟอล","แหล่ะ","ทวิต","แอคละ","เอนเกจ","กรวง","กาละเทศะ","แม่ม","คำพุดคำจา","แอ้บ","ถุกใจ","ไร้การพัฒนา",
    "กอลลัม","อิโมจิ","มึว","พ่อง","ชิหัย","คำถิ่น","รำคาน","นิวจีน","พ่อพระ",'กระทะทองแดง"',"ไบแอส","คร้า","ทุ่งลาเวนเดอร์","โซว์","แอค","ลิซ่า","วุ่นวายชีวิต","ประยุท",
    "คาแรคเต้อ","สมองกลวง","เซลสมอง","แอค","ฉิหาย","กระโหลกกะลา","ตอนคลอดผ่าน","กระหรี่","บอป","แบมแบม","เจฟอยู่","อารยะธรรม","เน็ตไอ","สันขวาน","ชิพหาย","ชาติละ","แคมโบ๋วเดียร์",
    "กะดอง","พิกกี้บู","มัมหมี","ฟัคยู","สูบพอต","บูลี่","สั่งสอง","พรรคส้ม","อีสัส","อีเหี้ย","อีเลว","อีดอก","อีเปรต","อีผีเปรต","ไอ้เหี้ย","เอ้มเน้ต","อีห่า","โดมะ","อีสัด","ไอช์","รุตัก","อีผีเปต","อีบ้า",
    "ไดโนเสา","อกแตกตาย","ไอ้สัส","ไอ้บิว","มูฟออน","อุบาต","สตอรี่","อับปรี","เอ้มวี","เหน","ยอด","กวนส้นตีน","ไอ้แม่เย็ด","เย็ดแม่","อัปปรีย์","ก๊อป","นางคนิกา",
    "อิเหี้ย","ไอ้ส้นตีน","ไอ้หน้าเหี้ย","โปรเม็ก","กัปปะ","แต้ด","อีควาย","หน้าส้นตีน","อีส้นตีน","ไอ้หน้าส้นตีน","อีนรก","ผู้หญิงต่ำ","อีสัตว์","ผีเปรต","หัวควย","โครต",
    "อีเดียร์","ตลลอด","อีควาย","แครป","อีหนังฮี๋","แม่เย็ด","เด็กเลี้ยงแกะ","อีโศกโครก","เย็ดเข้","ร่านกว่า","เยวหร้อก","โครต","เจรา","ดะดรีนาลีน","สุชาติ","เมเนเจอร์","เมเน",
    "เยดแม่","กําแพงเพชร","บ้านบึง","แล้วจะ","อาริ","จูจู","รุป","รูปอก","เสว","เห็นแล้วแข็ง","นำแตก","คุณอาเธอร์","ทุกรอบ","เช็กชี่","ไหญ่","ครัง","เห็นแล้ว","มอสส","น่าโดน","น้ำแตก",
    "น่าเอา","กระเเทกเเรง","เเตกใน","น่าล่อ","เนรตนารี","ลิมฝีปาก","แคท","ชุมแพ","ยินดีจริง","ออฟฟิสว่าง",
    "รุสึก","นอย","สมาย","ถูกเมิน","แค่นอย","รู้สึก","รูสึก","ไม่เป็นที่ยอมรับ","ไม่ถูกยอมรับ","ชนชั้นรากหญ้า","ไม่มีใครสนใจ",
    "ไม่เป็นที่รัก","ถูกกีดกัน","ถูกละเลย","ด้อยค่า","ไม่เท่าเทียม","ต่ำต้อย","ถูกลดคุณค่า","ถูกตัดขาด","ถูกตัดออก","โดนแยกออก","โดนมองข้าม","ไม่เท่าคนอื่น","ไอ้กระจอก","ไร้ตัวตน","คนละชั้น",
    "โดน","คนละครึ่ง","อีลีท","ซุปเปอร์โบล","เจษ","สีหศักดิ์","ศิริกัญญา","ศุภจี","สก้อย","ตบก","ช่วงนิ","ตุวเอง","ไม่ชอบ","อี๊ก","ตอนนิ้","คุนน้าาาา","โล้ด","เดว","ก้นอย",
    "พยสยาม","เต้นไม่เท่าคนอื่น","ผิวดี","เฉยม","เปนหมุเปนหมา","โคสอัพ","โดนทรีต","ชั้รรุ้สึก","เรีบน","เฟิสจ้อบเบ้อ","ทวิตละ","บิวตี้","ลีน","เเรก","คะแนนคงตก",
    "ถูกลดค่า","ถูกลดความสามารถ","บิวตี้สแตนดาร์ด","รู้สึก, แย่","ตุว","เรซูเม่","เหง๋า","ไมได","โดนเมิน","ไม่เป็นที่ยอมรับ", "ไม่ถูกยอมรับ",  "ถูกตัดขาด","แอนดรอย ",
    "อีนอกครอก","โหวตเตอร์","โดนรี","คนเฒ่า","อีลีต","อีลิท","ก็ับ","เด็็ก","เพศสภาพ","ไม้ร้อง","วนลูก","สวิสดิการ","สันดาร","แพนิค","ตายวะ","คอนเนคชั่น","ไทป์","ไม่เป้นที่รัก",
    "กลายเป็็นคน","ถูกลด","บูลลี่","เป็นสาวสอง","รุ้สึก","การทรีต","แอป","บางครั้้ง","ไม่มีค่า","ทำได้แค่นี้","สำคัญตัว","ไม่มีปัญญา","แลย","อุบาวท์","โคตรต่ำ","อิควาย","ภาระสังคม",
    "สมองน้อย","ส่ำส่อน","อีเวร","ประนาม","ไอเหี้ย","ไร้น้ำยา","ไร้ปัญญา","โคตรตระกลู",




]
# Force some words to split into multiple tokens
FORCE_SPLIT = {
    "เสนียดจัญไร": ["เสนียด","จัญไร"], "คนจัญไร":["คน","จัญไร"],"อัปรีย์จัญไร":["อัปรีย์","จัญไร"],
    "อิดอกทอง":["อี","ดอกทอง"], "แอคละ":["แอค","ละ"],"นัดเย็ด":["นัด","เย็ด"],"ตอนคลอดผ่าน":["ตอน","คลอด","ผ่าน"],
    "เจฟอยู่":["เจฟ","อยู่"],"น๊าาาา":["น๊า","าาา"],"ใช่ปะ":["ใช่","ปะ"],"ชาติละ":["ชาติ","ละ"],"เลวทรามต่ำช้า":["เลวทราม","ต่ำช้า"],
    "คนละโลก":["คน", "ละ", "โลก"],"จั๊กกก":["จั๊ก","กก"],"เกิ๊นนนนนนนนนนนนนนน":["เกิ๊น","นนนนนนนนนนนนนน"],
    "โสม":["โสม","ม"],"เยวหร้อก":["เยว","หร้อก"], "รุ้ก้":["รุ้","ก้"],"ขอดมหี":["ขอ","ดม","หี"], "เลียหี":["เลีย","หี"],
    "หี้ธอ":["หี","เธอ"],"หืมมมม":["หืม","มมม"],"น้ออออ":["น้อ","อออ"],"ชักนำ":["ชัก","นำ"], "คุณอาเธอร์":["คุณ","อาเธอร์"],
    "งื้ออออ":["งื้อ","อออ"],"มอสส":["มอส","ส"],"ยินดีจริง":["ยินดี","จริง"],"ออฟฟิสว่าง":["ออฟฟิส","ว่าง"],
    "ที่จะ":["ที่","จะ"],"ความต่ำต้อย":["ความ","ต่ำต้อย"],"แค่นอย":["แค่","นอย"],"ใจจริง":["ใจ","จริง"],"ถุกกก":["ถูก","กก"],
    "คุนน้าาาา":["คุนน้า","าาา"],"ก้นอย":["ก็","นอย"],"มมาก":["ม","มาก"],"เฉยม":["เฉย","ม"],"เปนหมุเปนหมา":["เปนหมุ","เปนหมา"],
    "ชั้รรุ้สึก":["ฉัน","รู้สึก"],"ทวิตละ":["ทวิต","ละ"],"คะแนนคงตก":["คะแนน","คง","ตก"], "แม่งแย่ชห":["แม่งแย่","ฉิบหาย"],"เพื่อนทรีต":["เพื่อน","ทรีต"],
    "ตายวะ":["ตาย","วะ"],"กลายเป็็นคน":["กลายเป็น","คน"],"เป็นสาวสอง":["เป็น","สาวสอง"]
}
# Normalize slang, typo, and informal words
FIX_TOKENS = {
    "กอ": "กอด", "บอ":"บอก", "ชม":"ชั่วโมง", "กี":"หี",
    "ฝรั่งเศษ":"ฝรั่งเศส", "สว":"สมาชิกวุฒิสภา", "เหรี้ยย":"เหี้ย",
    "เค้า":"เขา","รพ":"โรงพยาบาล","เดะ":"เดี๋ยว","เหมน":"เขมร","แว้น":"ขับขี่",
    "มรึง":"มึง","สถุน":"สถุล","พน":"พรุ่งนี้","สน":"สถานีตำรวจ","สส":"สมาชิกสภาผู้แทนราษฎร",
    "ก้อ":"ก็","อุบาด":"อุบาทว์","ปท":"ประเทศ","กุ":"กู","มุง":"มึง","วา":"อันวาร์",
    "ปห": "ปัญหา","โด": "โดด","มก": "มาก","ต้ำ": "ต่ำ","กักขระ": "อักขระ","แอคหลุม": "บัญชีสำรอง",
    "จังไร":"จัญไร","แตด":"หี","อิสัส":"อีสัส","เย้สเขร้":"เย็ดเข้","NGO":"โง่","tadddd":"หี","อัปปรีย์":"อัปรีย์","อส":"อีสัส","อีสัด":"อีสัส",
    "เสล่อ":"สะเหล่อ","เสร่อ":"สะเหล่อ", "บัง":"ยัง","สะ":"ซะ","อิ":"อี","ยาด":"อนุญาต",
    "ชิปฝาย":"ฉิบหาย","เมกา":"อเมริกา","เป้น":"เป็น","ต้าว":"ที่รัก",
    "ฟอล":"ติดตาม", "ไอ่":"ไอ้", "โพส":"โพสต์", "ซัง":"เกลียด", "ทวิต":"ทวิตเตอร์", "แอค":"บัญชี", "กรวง":"กลวง", "ปชช":"แปลว่า ประชาชน",
    "ง้าว":"โง่", "แม่ม":"แม่ง", "คำพุดคำจา":"คำพูดคำจา", "แอ้บ":"เสแสร้ง", "ถุกใจ":"ถูกใจ", "ก้":"ก็","มึว":"มึง", "พ่อง":"พ่อ", "ชิหัย":"ชิบหาย",
    "หุ":"หู", "รำคาน":"รำคาญ", "เมิง":"มึง","จั้ด":"ตัดสิน","เน้ต":"เน็ต","แปป":"สักครู่","กะทะ":"กระทะ","กาละเทศะ":"กาลเทศะ",
    "งัย":"ไง","นย":"นัดเย็ด","เกิ๊นน":"เกิน","ทร":"ทรมาน","เชี่ย":"เหี้ย","คร้า":"ค่ะ","สัด":"สัส","บอกอ":"บอก","ฟค":"แฟนคลับ","โซว์":"โชว์",
    "เอว":"เอง","เลน":"เลย","โขมย":"ขโมย","หยุ":"หยุด","ประยุท":"ประยุทธ์","ชิพหาย":"ฉิบหาย","ดาหน้า":"ตราหน้า","มั่ย":"ไหม","เปน":"เป็น", "เซลสมอง":"เซลล์สมอง", "ฉิหาย":"ฉิบหาย",
    "ชิบหาย":"ฉิบหาย", "โง๊":"โง่", "ค":"ควาย", "จิง":"จริง", "กระหรี่":"กะหรี่", "บอป":"ปอบ","กระโหลกกะลา":"กะโหลกกะลา","แบ":"แบม",
    "กทม":"กรุงเทพมหานคร", "ไอ":"ไอ้", "กรีด":"กรี๊ด", "อารยะธรรม":"อารยธรรม", "เน็ตไอ":"เน็ตไอดอล", "ร่่น":"ร่าน", "จื๋ม":"หี","ยุ":"อยู่","บัก":"ไอ้","ปั๊ค":"ปลั๊ก","ส่า":"ว่า",
    "แคมโบ๋วเดียร์":"กัมพูชา","ฟี":"ความรู้สึก","กะดอง":"ควย","กะดอ":"ควย","ตะหาก":"ต่างหาก","ศลป":"ศิลปิน","ฟัคยู":"ควย",
    "แอ็ค":"บัญชี", "สูบพอต":"สูบบุหรี่ไฟฟ้า", "บูลี่":"บูลลี่", "สั่งสอง":"สั่งสอน", "ตุ้ด":"กะเทย", "โดดมะ":"โดมะ","เจ๋ง":"เจ๊ง",
    "ไอช์":"ไอซ์", "กะ":"กับ", "รุตัก":"รู้จัก", "อีผีเปต":"อีผีเปรต", "ไดโนเสา":"ไดโนเสาร์","อุบาท":"อุบาทว์","อุบาต":"อุบาทว์","ค๊วย":"ควย","อับปรี":"อัปปรีย์","เอ้มวี":"เอ็มวี","แระ":"ละ","เหน":"เห็น","รุ้":"รู้",
    "แดรก":"แดก","อัปปรีย์":"อัปปรีย์","เกิ๊น":"เกิน","แต้ด":"หี","อีสัตว์":"อีสัส","ตลลอด":"ตลอด","ตัง":"ตังค์","ม๊าก":"มาก","แครป":"บันทึก","อีหนังฮี๋":"อีหนังหี",
    "อีโศกโครก":"อีโสโครก","ใด้":"ได้","เยว":"เดียว","หร้อก":"หรอก","ใอ้":"ไอ้","ปทท":"ประเทศ","โครต":"โคตร","เจรา":"เจรจา","ดะดรีนาลีน":"อะดรีนาลีน","ก้":"ก็",
    "เท":"เธอ","จู๋":"ควย","หรรม":"ควย","ตอก":"เย็ด","เด้า":"เย็ด","เย":"เย็ด","เย้":"เย็ด","กุ่น":"หุ่น","บาก":"อยาก","็ด":"เย็ด","น้ง":"น้อง","นอ้ง":"น้อง","อ้ง":"น้อง","เสว":"เสียว","นำแตก":"น้ำแตก",
    "ไหญ่":"ใหญ่","รู่":"รู้","ครัง":"ครั้ง","เซ็กส์":"เย็ด","นู๋":"หนู","สี้":"เย็ด","รุสึก":"รู้สึก","ได":"ได้","ยเาย":"ย้าย","รูสึก":"รู้สึก","ถุก":"ถูก","ตบก":"ตลก",
    "ช่วงนิ":"ช่วงนี้","ตุวเอง":"ตัวเอง","อี๊ก":"อีก","ตอนนิ้":"ตอนนี้","คุนน้า":"คุณน้า","ชั้น":"ฉัน","เดว":"เดียว","พยสยาม":"พยายาม","ทัง":"ทั้ง","กัร":"กัน",
    "เรีบน":"เรียน","เสา":"เสาร์","ทิต":"ทิตย์","จา":"จะ","ชห":"ฉิบหาย","ตุว":"ตัว","เหง๋า":"เหงา","ใคา":"ใคร","ไ่ม่":"ไม่","ไมได":"ไม่ได้","หมพ":"หมด","ก็ับ":"กับ","เด็็ก":"เด็ก",
    "ไม้ร้อง":"ไม่ร้อง","มี่":"นี่","วนลูก":"วนลูป","สวิสดิการ":"สวัสดิการ","สันดาร":"สันดาน","ไม่เป้นที่รัก":"ไม่เป็นที่รัก","รุู้":"รู้","เป็็นคน":"เป็นคน","รุ้สึก":"รู้สึก","แลย":"เลย",
    "บางครั้้ง":"บางครั้ง","อุบาวท์":"อุบาทว์","ประนาม":"ประณาม","ไอเหี้ย":"ไอ้เหี้ย"
}
# Filler/noise words to remove
FILLER_WORDS = {
    # เสียงอุทาน
    "อะ","อ่ะ","เออ","เอ่อ","อืม","อือ","อ่า","อ่อ","โอย","อ้ะ","เอ๊ย",
    # คำลงท้าย
    "นะ","น้า","นะคะ","นะครับ","สิ","ซิ","ล่ะ","ละ",
    "แหละ","แหล่ะ","เหอะ","วะ","หว่ะ","มั้ง","เนี้ย","อ้อ","นี้แหละ","เหรอ","โน่น","แบบนี้","ซะ","ว่ะ","ดิ","'งี้",
    "น่ะ","แง","น๊า","อุ้ย","ป่ะ","หวะ","น๊ะ","งื้อ",

    # สุภาพ
    "ครับ","ครับผม","ค่ะ","คะ","จ้า","จ๊ะ","จ๋า",
    "ขอรับ","เพคะ",

    # คำเชื่อม/ฟิลเลอร์
    "ก็","เลย","แบบ","แบบว่า","คือ","คือว่า","ที่","งือ","แบบนี้",

    # คำขยายฟุ่มเฟือย
    "จริงๆ","จริงจริง","โคตรจะ","อย่างมาก","อย่างโคตร",
    "เอ้ย","โถ","แหล่ะ","เอ๊ะ","ละ","ไใ่","หรอ","แหม","อะดิ","ปะ","ง่ะ","เว่อ","โอ๊ย","เนี้ยะ","า",
    "ป้ะ","เนาะ","โถ่","เรอะ","เออ","นุง","นะจ๊ะ","ว้าว","เว้ย","หรือว่า","รึเปล่า","หนิ","หว้า","เก่","โอ้ย","ปส","งี้","อิอิ","อ๊ะ",
    "หว่า","เห้อ","ห้ะ","น้อ","เนอะ","ดิ่","อ๋า","คระ","ฮะ"


}

# Merge multi-word phrases into one token
PHRASE_MERGE = {
    ("เห็นแล้ว","แข็ง"):"เห็นแล้วแข็ง", ("ลูบ","คลำ"): "ลูบคลำ",("ถูก","เมิน"): "ถูกเมิน"

}

# Replace slang/typos with normalized words
def fix_tokens(tokens):
    return [FIX_TOKENS.get(t, t) for t in tokens]
# Force split selected words
def force_split(tokens):
    out = []
    for t in tokens:
        if t in FORCE_SPLIT:
            out.extend(FORCE_SPLIT[t])
        else:
            out.append(t)
    return out

# Reduce repeated characters
# Example: "อืมมม" -> "อืม"
def normalize_filler(token: str) -> str:
    token = re.sub(r"([ก-๙])\1{1,}", r"\1", token)
    return token

# Remove filler/noise words
def remove_filler_tokens(tokens):
    out = []
    for t in tokens:
        norm = normalize_filler(t)
        if norm not in FILLER_WORDS:
            out.append(t)
    return out

# Regex for repeated character tokens
REPEAT_CHAR_TOKEN = re.compile(r"^([ก-๙A-Za-z])\1+$")

# Remove tokens like "55555", "กกกก"
def drop_repeat_char_tokens(tokens):
    return [t for t in tokens if not REPEAT_CHAR_TOKEN.match(t)]

# Build custom dictionary with Thai words + custom words
trie = Trie(set(thai_words()).union(custom_words))


# Main preprocessing function
def split_word(text):
    text = str(text)
    # Tokenize text using custom dictionary
    tokens = word_tokenize(
        text,
        engine="newmm",
        custom_dict=trie,
        keep_whitespace=False
    )
    # Merge phrase tokens
    def merge_phrases(tokens):
        out = []
        i = 0
        while i < len(tokens):
            # merge 3 tokens
            if i+2 < len(tokens) and (tokens[i], tokens[i+1], tokens[i+2]) in PHRASE_MERGE:
                out.append(PHRASE_MERGE[(tokens[i], tokens[i+1], tokens[i+2])])
                i += 3
                continue

            # merge 2 tokens (ของเดิม)
            if i+1 < len(tokens) and (tokens[i], tokens[i+1]) in PHRASE_MERGE:
                out.append(PHRASE_MERGE[(tokens[i], tokens[i+1])])
                i += 2
                continue

            out.append(tokens[i])
            i += 1
        return out




    # Regex for English characters
    ENG_RE = re.compile(r"[A-Za-z]")
    # Tokens to remove
    REMOVE_TOKENS = {"อา", "าาร์"}
    # Normalize words
    tokens = fix_tokens(tokens)
    # Remove empty spaces
    tokens = [t for t in tokens if t.strip()]
    # Merge phrases
    tokens = merge_phrases(tokens)
    # Force split tokens
    tokens = force_split(tokens)
    # Remove stopwords
    tokens = [t for t in tokens if (t not in th_stop and t not in en_stop)]
    # Remove numbers
    tokens = [i for i in tokens if not i.isnumeric()]
    # Remove English tokens
    tokens = [t for t in tokens if not ENG_RE.search(t)]
    # Remove filler words
    tokens = remove_filler_tokens(tokens)
    # Remove repeated character tokens
    tokens = drop_repeat_char_tokens(tokens)
    # Remove short tokens
    tokens = [t for t in tokens if len(t) > 1]
    # Remove unwanted tokens
    tokens = [t for t in tokens if t not in REMOVE_TOKENS]


    return tokens

# Apply preprocessing to dataset
df_all["tokens"] = df_all["Cleaned_Text"].apply(split_word)
# Display results
df_all[["Cleaned_Text", "tokens","Label" ]].head(2100)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Convert token lists into text strings
texts = df_all['tokens'].apply(lambda x: ' '.join(x))

# 2. Create TF-IDF feature matrix
vectorizer_all = TfidfVectorizer(tokenizer=lambda x: x.split(), token_pattern=None)
X_tfidf_all = vectorizer_all.fit_transform(texts)

# 3. Convert sparse matrix into DataFrame
tfidf_all_df = pd.DataFrame(X_tfidf_all.toarray(), columns=vectorizer_all.get_feature_names_out())

# 4. Add Label column
tfidf_all_df['Label'] = df_all['Label'].values

# Display sample results
print(f"จำนวน Feature ทั้งหมด: {len(vectorizer_all.get_feature_names_out())} คำ")
display(tfidf_all_df.head())
```
